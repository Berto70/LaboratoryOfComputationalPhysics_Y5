2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
with  open('/Users/gabri/05_data/credit_card.dat','rb') as file:
    
    for line in file.readlines():
        char_size = 6
        for i in range(0, len(line)-char_size, char_size):
            print(chr(int(line[i:i+char_size], 2)), end= '')
        print('')
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
import json
data = json.load(open('/Users/gabri/05_data/user_data.json'))
print (data)

with  open('/Users/gabri/05_data/user_data.csv', mode='w') as f:
    filtered_data=[ele for ele in data if (ele['CreditCardType']=='American Express')]
    key = filtered_data[0].keys()
    f.write(', '.join(key)+'\n')
    for p in filtered_data:
        for k in list(p.keys())[:-1]:
            f.writelines(p[k]+', ')
        f.write((p[list(p.keys())[-1]]) + '\n')
```

6\. Load the file from this url: [https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1](https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1) with Pandas. 
+ Explore the data (see the info of the data)
+ Draw the istogram of the 'class' field. Decribe wath yuou see


```python
import matplotlib.pyplot as plt

url = "https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1"
data = pd.read_csv(url)

print('Data info:\n')
pd.DataFrame.info(data)
a=pd.DataFrame(data)

data


#plot
fig, ax = plt.subplots(nrows= 1, ncols=1, figsize=(5, 5))
ax.hist(a['class'])
ax.minorticks_on
ax.set_title('Class', fontsize=15)
ax.grid(visible=True)
fig.tight_layout()

#quickier method
data.hist("class")
```

7\. Load the remote file [https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1](https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1) with Pandas and plot a scatter plot all possible combination of the following fields:
    
  + features_1
  + features_2
  + features_3

```python
import matplotlib.pyplot as plt

url = "https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1"
data = pd.read_csv(url)
data

fig, ax = plt.subplots(nrows= 1, ncols=3, figsize=(20, 6))
ax[0].scatter(data['features_1'],data['features_2'])
ax[1].scatter(data['features_1'],data['features_3'])
ax[2].scatter(data['features_2'],data['features_3'])

ax[0].set_title("1&2", fontsize=15)
ax[1].set_title("1&3", fontsize=15)
ax[2].set_title("2&3", fontsize=15)
```

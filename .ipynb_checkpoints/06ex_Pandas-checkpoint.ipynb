## Pandas analysis

This exercise consists in analyzing a dataset containg timing information from a series of Time-to-Digital-Converters (TDC) implemented in a couple of FPGAs. Each measurement (i.e. each row of the input file) consists of a flag that specifies the type of message ('HEAD', which in this case is always 1), two addresses of the TDC providing the signal ('FPGA' and 'TDC_CHANNEL'), and the timing information ('ORBIT_CNT', 'BX_COUNTER', and 'TDC_MEAS'). Each TDC count corresponds to 25/30 ns, whereas a unit of BX_COUNTER corresponds to 25 ns, and the ORBIT_CNT is increased every 'x' BX_COUNTER. This allows to store the time in a similar way to hours, minutes and seconds.

1\. Create a Pandas DataFrame reading N rows of the 'data_000637.txt' dataset. Choose N to be smaller than or equal to the maximum number of rows and larger that 10k.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_name="06_data/data_000637.txt"
dataF = pd.DataFrame(pd.read_csv(file_name))
dataF
```

2\. Find out the number of BX in a ORBIT (the value 'x').

```python
#il valore di orbit è aumentato di 1 ogni x bx_counter.
#quando bx counter arriva a x il valore di orbit è aumentato di 1
#cerco il valore massimo e aggiungo 1, così so dopo quanti conteggi aumenta l'orbit

x = dataF['BX_COUNTER'].max() + 1

#print(dataF['BX_COUNTER'].max())
#print(dataF['BX_COUNTER'].min())

print("The ORBIT_CNT is increased every ", x ," BX_COUNTER")
```

3\. Find out how much the data taking lasted. You can either make an estimate based on the fraction of the measurements (rows) you read, or perform this check precisely by reading out the whole dataset.

```python
time_ns = (dataF["ORBIT_CNT"].max() - dataF["ORBIT_CNT"].min())*25*x
#print (diff)

time_s = time_ns/(10e9)

print("Estimated elapsed time:", time_s, "sec")

N = dataF.shape[0]

dataF = dataF.sort_values(['ORBIT_CNT', 'BX_COUNTER', 'TDC_MEAS'], ignore_index=True)
time0 = (dataF['BX_COUNTER'][0])*25 + (dataF['TDC_MEAS'][0])*25/30 + (dataF['ORBIT_CNT'][0])*25*x
timef = (dataF['BX_COUNTER'][N-1])*25 + (dataF['TDC_MEAS'][N-1])*25/30 + (dataF['ORBIT_CNT'][N-1])*25*x

precise_time_s = (timef - time0)/(10e9)

print(time0)

print("Precise elapsed time:", precise_time_s, "sec")

print("relative % error:", round(abs(time_s - precise_time_s)/precise_time_s*100, 4),"%")
```

4\. Create a new column with the absolute time in ns (as a combination of the other three columns with timing information).

```python
dataF["ABS_TIME"] = dataF['BX_COUNTER']*25 + dataF['TDC_MEAS']*25/30 + dataF['ORBIT_CNT']*25*x

dataF["ABS_TIME"] = (dataF['ORBIT_CNT']-(dataF['ORBIT_CNT'][0]))*x*25 + (dataF['BX_COUNTER']-(dataF['BX_COUNTER'][0]))*25 + (dataF['TDC_MEAS']-(dataF['TDC_MEAS'][0]))*(25/30)
```
